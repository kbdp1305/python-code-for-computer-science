# Taking Max Value Only

```python
Topic_Grouped = [defaultdict(lambda: [0, 0]) for _ in range(len(tokenized_docs))]
```

```python
for i,j in enumerate(probability_dict) :
  for keys in j.keys() :
    value=probability_dict[i][keys]
    max_index = value.index(max(value))
    Topic_Grouped[i][keys]=max_index
```

```python
# Mapping idx to words
def replace_keys_with_words_in_topic_group(topic_group, id_to_word):
    replaced_topic_group = []
    for defaultdict_item in topic_group:
        replaced_defaultdict = defaultdict(lambda: [0, 0])
        for key, value in defaultdict_item.items():
            replaced_key = id_to_word.get(key, key)
            replaced_defaultdict[replaced_key] = value
        replaced_topic_group.append(replaced_defaultdict)
    return replaced_topic_group

# Mapping keys from id to word
replaced_Topic_Grouped = replace_keys_with_words_in_topic_group(Topic_Grouped, id_to_word)
```
